## Carga de paquetes, funciones y dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
def linear_cost (X,
                 y,
                 theta,
                 la_lambda):
    return ((np.matmul(X, theta)-y)**2).sum()*(1/2*X.shape[0])  + (theta ** 2).sum()*(la_lambda/2*X.shape[0])

def linear_cost_derivate(X,
                         y,
                         theta,
                         la_lambda):
    return np.matmul((np.matmul(X, theta)-y).T,X).T.sum()/X.shape[0] + theta.sum()*la_lambda/X.shape[0]

def gradient_descent(
    X,
    y,
    theta_0,
    cost,
    cost_derivate,
    alpha=0.001,
    la_lambda=0.0001,
    treshold=0.0001,
    max_iter=10000):
    costs = []
    gradient_norms = []
    theta, i = theta_0, 0
    while np.linalg.norm(cost_derivate(X,y,theta,alpha)) > treshold and i <max_iter:
        theta = theta - (alpha * cost_derivate(X,y,theta,alpha))
        i +=1
        costs.append(cost(X, y, theta,alpha))
        gradient_norms.append(cost_derivate(X, y, theta, alpha))
    return theta, costs, gradient_norms
```

## Ejemplos en clase

```python
def linear_cost_derivate(X, y, theta):
    h = np.matmul(X, theta)
    m, _ = X.shape
    return np.matmul((h - y).T, X).T.sum() / m

def linear_cost(X, y, theta):
    m, _ = X.shape
    h = np.matmul(X, theta)
    sq = (y - h) ** 2
    return sq.sum() / (2 * m)

def gradient_descent(
        X,
        y,
        theta_0,
        cost,
        cost_derivate,
        alpha=0.01,
        treshold=0.0001,
        max_iter=10000):
    theta, i = theta_0, 0
    costs = []
    gradient_norms = []
    while np.linalg.norm(cost_derivate(X, y, theta)) > treshold and i < max_iter:
        theta =  theta - alpha * cost_derivate(X, y, theta)
        i += 1
        costs.append(cost(X, y, theta))
        gradient_norms.append(cost_derivate(X, y, theta))
    return theta, costs, gradient_norms

```
